```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataset
data = pd.read_csv('/content/titanic.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# Drop rows where the 'Embarked' column has missing values
data.dropna(subset=['Embarked'], inplace=True)
# Fill missing values in the 'Cabin' column with the string "Unknown"
data["Cabin"].fillna("Unknown", inplace = True)
# Fill missing values in the 'Age' column with the mean age
data["Age"].fillna(data["Age"].mean(), inplace = True)
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(6, 3))
sns.set_style("whitegrid")
sns.histplot(data["Age"], kde=True,color="skyblue")
plt.title("Age vs no of people",fontsize=16)
plt.xlabel("Age",fontsize=14)
plt.ylabel("no of people",fontsize=14)
plt.show()
```

```python
# Plot the count of survivors and non-survivors, separated by gender
plt.figure(figsize=(6, 3))
sns.set_style("white")
sns.countplot(data=data, x="Sex", hue="Survived", palette="pastel", edgecolor=".6")
plt.title("no of people alive by Gender", fontsize=18, weight='bold')
plt.xlabel("Gender")
plt.ylabel("no of pepole ")
plt.legend(title="Survived", loc="upper right", fontsize=12, title_fontsize='14')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Plot a line chart of 'Age' vs 'Fare', colored by survival status
sns.set_style("whitegrid")
plt.figure(figsize=(6, 3))
sns.lineplot(data=data, x="Age", y="Fare", hue="Survived", marker="o", style="Survived", dashes=False)
plt.title("Line Chart of Age vs Fare", fontsize=16, weight='bold')
plt.xlabel("Age", fontsize=14)
plt.ylabel("Fare", fontsize=14)
plt.legend(title="Survived", loc="upper right")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
